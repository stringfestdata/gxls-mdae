```python
import openpyxl
from openpyxl.chart import ScatterChart, Reference, Series

# Create a new workbook
workbook = openpyxl.Workbook()

# Select the active sheet
sheet = workbook.active

# Sample data
data = [
    (1, 3),
    (2, 5),
    (3, 4),
    (4, 6),
    (5, 8),
]

# Write the data to the worksheet
for row in data:
    sheet.append(row)

# Create a scatter chart
chart = ScatterChart()

# Set the x-axis values
xvalues = Reference(sheet, min_col=1, min_row=1, max_row=len(data))

# Set the y-axis values
yvalues = Reference(sheet, min_col=2, min_row=1, max_row=len(data))

# Create a series with x and y values
series = Series(yvalues, xvalues, title="Sample Data")

# Set the marker style to 'dot'
series.marker.symbol = "dot"

# Set the line style to 'none'
series.graphicalProperties.line.noFill = True

# Add the series to the chart
chart.series.append(series)

# Set the title of the x-axis
chart.x_axis.title = "X Values"

# Set the title of the y-axis
chart.y_axis.title = "Y Values"

# Remove gridlines
chart.x_axis.majorGridlines = None
chart.y_axis.majorGridlines = None

# Add the chart to the worksheet
sheet.add_chart(chart, "D2")

# Save the workbook
workbook.save("scatterplot.xlsx")
```
